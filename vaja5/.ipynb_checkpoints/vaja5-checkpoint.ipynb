# MULTIPLA REGRESIJA

```python
import pandas as pd
import numpy as np

df = pd.read_excel("Izdelava3DPrototipa_Baza_SKUPINE_pripravljena.xlsx")

# --- USTVARIMO KATEGORIČNE SPREMENLJIVKE ---
# 1) Trajanje_kategorija
df['Trajanje_kategorija'] = pd.cut(
    df['Trajanje_h'],
    bins=[0, 1, 2.5, 10],
    labels=['Kratko', 'Srednje', 'Dolgo']
)

# 2) Skupina glede na VA / NVA / NNVA
df['Skupina'] = df['Tip_aktivnosti'].apply(lambda x: 'VA' if x == 'VA' else 'NVA_NNVA')

# PREVERIMO MANJKAJOČE PODATKE
#print(df.isna().sum())
```

## Model 1: Enostavna linearna regresija

Cilj Modela 1 je z linearno regresijo preveriti, ali obstaja statistično značilna linearna povezava med trajanjem koraka in časom čakanja v procesu. Ta analiza spada v fazo Analyze pristopa Lean Six Sigma, kjer identificiramo potencialne $X$ vplivne dejavnike na $Y$ izid.

Za ta model smo uporabili naslednje spremenljivke iz podatkovne baze:

Ciljna spremenljivka ($Y$): Čakanje_h
 * Predstavlja čas čakanja (v urah) na določeni točki v procesu. V kontekstu procesa ta spremenljivka predstavlja izgubo (Muda) in je ključna metrika, ki jo želimo zmanjšati. Je zvezna spremenljivka.

Napovedna spremenljivka ($X_1$): Trajanje_h
* Predstavlja dejansko trajanje (v urah) enega določenega koraka v procesu. Služi kot preprost napovednik, da ugotovimo, ali so daljši koraki povezani z daljšim ali krajšim časom čakanja. Je zvezna spremenljivka.




```python
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error

plt.style.use("dark_background")  # dark theme za vse grafe

# Priprava podatkov
df_lin = df[['Trajanje_h', 'Čakanje_h']].dropna()

X = df_lin[['Trajanje_h']]
y = df_lin['Čakanje_h']

# dodamo konstanto za OLS (intercept)
X_const = sm.add_constant(X)

# Gradnja modela (OLS linearna regresija)
model1 = sm.OLS(y, X_const).fit()
print(model1.summary())

# osnovne metrike na TRAIN podatkih
y_pred = model1.predict(X_const)
residuals = y - y_pred

rmse = np.sqrt(mean_squared_error(y, y_pred))
mae = mean_absolute_error(y, y_pred)
r2_train = model1.rsquared

print(f"\nTrain metrike za Model 1:")
print(f"R² (train) = {r2_train:.3f}")
print(f"RMSE (train) = {rmse:.3f}")
print(f"MAE  (train) = {mae:.3f}")

#10-fold cross-validation (R²)
X_cv = df_lin[['Trajanje_h']].values
y_cv = df_lin['Čakanje_h'].values

kf = KFold(n_splits=10, shuffle=True, random_state=42)
lin_reg = LinearRegression()

cv_scores = cross_val_score(lin_reg, X_cv, y_cv, cv=kf, scoring='r2')

print(f"\n10-fold CV R²: {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")

#Graf 1: Čakanje vs Trajanje + regresijska premica
plt.figure(figsize=(6,4))
plt.scatter(df_lin['Trajanje_h'], df_lin['Čakanje_h'],
            alpha=0.7, label='Opazovanja', color="deepskyblue")

x_line = np.linspace(df_lin['Trajanje_h'].min(), df_lin['Trajanje_h'].max(), 100)
y_line = model1.params['const'] + model1.params['Trajanje_h'] * x_line
plt.plot(x_line, y_line, color='red', label='Regresijska premica')

plt.title('Model 1: Čakanje v odvisnosti od trajanja koraka')
plt.xlabel('Trajanje (h)')
plt.ylabel('Čakanje (h)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("Komentar: graf prikazuje posamezna opazovanja in linearno regresijsko premico.")

#Graf 2: Dejanske vs napovedane vrednosti
plt.figure(figsize=(6,4))
plt.scatter(y_pred, y, alpha=0.7, color="orange", label="Opazovanja")
min_val = min(y.min(), y_pred.min())
max_val = max(y.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val],
         color="white", linestyle='--', label="Idealna (y = y_hat)")

plt.title('Model 1: Dejanske vs napovedane vrednosti')
plt.xlabel('Napovedano čakanje (h)')
plt.ylabel('Dejansko čakanje (h)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("Komentar: bližje kot so točke diagonali, bolj natančne so napovedi modela.")

#Graf 3: Ostanki (residuals) vs napovedane vrednosti
plt.figure(figsize=(6,4))
plt.scatter(y_pred, residuals, alpha=0.7, color="lime")
plt.axhline(0, color='white', linestyle='--')

plt.title('Model 1: Ostanki glede na napovedane vrednosti')
plt.xlabel('Napovedano čakanje (h)')
plt.ylabel('Ostanek (residual)')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("Komentar: ostanki bi morali biti približno naključno razpršeni okoli 0 (brez vzorca).")

#Graf 4: Pomembnost spremenljivke (koeficient)
coef_names = ['Trajanje_h']
coef_values = [model1.params['Trajanje_h']]

plt.figure(figsize=(4,4))
plt.bar(coef_names, coef_values, color="gold")
plt.axhline(0, color='white', linestyle='--')

plt.title('Model 1: Koeficient regresije')
plt.ylabel('β (ne-standardiziran)')
plt.tight_layout()
plt.show()

print("Komentar: negativni koeficient pomeni, da daljše trajanje zmanjšuje čakanje.")

#Graf 5: 10-fold cross-validation (R²)
plt.figure(figsize=(6,4))

# boxplot R² po foldih
plt.boxplot(cv_scores, vert=True)
plt.title('Model 1: 10-fold cross-validation (R²)')
plt.ylabel('R²')
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.bar(np.arange(1, 11), cv_scores, color="dodgerblue")
plt.title('Model 1: R² po posameznih foldih (10-fold CV)')
plt.xlabel('Fold')
plt.ylabel('R²')
plt.tight_layout()
plt.show()

print("Komentar: boxplot in stolpčni graf prikazujeta stabilnost R² med posameznimi foldi.")
```

Ali ima trajanje vpliv na čakanje? Trajanje_h β = –0.4133, p < 0.001 (Če se trajanje poveča za 1h se čakanje v povprečju zmanjša za 0.41 h

Kako dobro model pojasni čakanje? R² = 0.392 Model pojasni 39.2 % variabilnosti čakanja.
10-fold CV R² = 0.281 ± 0.255 Model je zmeren in nekoliko nestabilen,


## MODEL 2 - Multilinearna linearna regresija

Cilj Modela 2 je nadgraditi in izboljšati Model 1 z vključitvijo več napovednih spremenljivk (multivariatna linearna regresija). Model preverja, kako se pojasnjevalna moč in napovedna natančnost spremenita, če poleg trajanja koraka vključimo še dejavnika, povezana z variabilnostjo procesa in kompleksnostjo izvedbe. Ta model bo omogočil tudi oceno relativne pomembnosti vsakega napovednika na izid $Y$ (s pomočjo standardiziranih koeficientov).

Definicija spremenljivkOhranimo isto ciljno spremenljivko $Y$ kot v Modelu 1, vendar dodamo nove napovednike $X$:
* Ciljna spremenljivka ($Y$): Čakanje_hOpis: Čas čakanja v procesu (ura).
* Napovedna spremenljivka ($X_1$): Trajanje_hOpis: Dejansko trajanje koraka (ura). (Ohranjena iz Modela 1)
* Napovedna spremenljivka ($X_2$): NapakeOpis: Število napak ali popravkov, ki so bili potrebni v tem koraku. Pričakujemo, da večje število napak pomeni večjo variabilnost in s tem daljše čakanje.
* Napovedna spremenljivka ($X_3$): IteracijaOpis: Število ponovitev določenega koraka. To je kazalnik kompleksnosti ali nestabilnosti. Pričakujemo, da več iteracij pomeni daljše čakanje.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# 2) Multilinearna regresija: Čakanje_h ~ Trajanje_h + Napake + Iteracija

# Poskrbimo, da imamo numerične stolpce
df_lin2 = df[['Čakanje_h', 'Trajanje_h', 'Napake', 'Iteracija']].dropna().copy()

y2 = df_lin2['Čakanje_h']
X2 = df_lin2[['Trajanje_h', 'Napake', 'Iteracija']]

# Dodamo konstanto
X2_const = sm.add_constant(X2)

model2 = sm.OLS(y2, X2_const).fit()
print(model2.summary())

# Standardizirani koeficienti (beta)
y_std = y2.std()
beta_std = {}
for col in X2.columns:
    beta_std[col] = model2.params[col] * (X2[col].std() / y_std)

print("\nStandardizirani koeficienti (β_std):")
for col, val in beta_std.items():
    print(f"{col}: {val:.3f}")

# 10-fold cross-validation za R², RMSE, MAE
X2_cv = X2.values
y2_cv = y2.values
kf = KFold(n_splits=10, shuffle=True, random_state=42)
lin2 = LinearRegression()

cv_r2 = cross_val_score(lin2, X2_cv, y2_cv, cv=kf, scoring='r2')
cv_rmse = -cross_val_score(lin2, X2_cv, y2_cv, cv=kf, scoring='neg_root_mean_squared_error')
cv_mae = -cross_val_score(lin2, X2_cv, y2_cv, cv=kf, scoring='neg_mean_absolute_error')

print(f"\n10-fold CV R²: {cv_r2.mean():.3f} ± {cv_r2.std():.3f}")
print(f"10-fold CV RMSE: {cv_rmse.mean():.3f} ± {cv_rmse.std():.3f}")
print(f"10-fold CV MAE: {cv_mae.mean():.3f} ± {cv_mae.std():.3f}")

# Graf: dejansko vs napovedano
y2_pred = model2.predict(X2_const)

plt.style.use('dark_background')
plt.figure(figsize=(6, 4))
plt.scatter(y2, y2_pred, alpha=0.7)
plt.plot([y2.min(), y2.max()], [y2.min(), y2.max()], 'r--')
plt.title('Model 2: Dejanske vs napovedane vrednosti (Čakanje_h)')
plt.xlabel('Dejanske vrednosti')
plt.ylabel('Napovedane vrednosti')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Graf: ostanki
resid2 = y2 - y2_pred

plt.figure(figsize=(6, 4))
plt.scatter(y2_pred, resid2, alpha=0.7)
plt.axhline(0, color='r', linestyle='--')
plt.title('Model 2: Ostanki glede na napovedi')
plt.xlabel('Napovedane vrednosti')
plt.ylabel('Ostanki')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Coefficient plot (pomembnost spremenljivk)
coef_vals = model2.params.drop('const')

plt.figure(figsize=(6, 4))
coef_vals.plot(kind='bar')
plt.title('Model 2: Koeficienti napovednih spremenljivk')
plt.ylabel('β')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Boxplot cross-validation R²
plt.figure(figsize=(5, 4))
plt.boxplot(cv_r2, vert=True)
plt.title('Model 2: 10-fold CV R²')
plt.ylabel('R²')
plt.tight_layout()
plt.show()
```

## MODEL 3 – Logistična regresija
Cilj Modela 3 je preiti iz napovedovanja zvezne metrike ($Y = \text{'Čakanje\_h'}$) na problem klasifikacije. Uporabljamo logistično regresijo, da preverimo, kateri dejavniki procesa pomembno vplivajo na verjetnost nastanka vsaj ene napake v določenem koraku. Ta sprememba je ključna, saj logistična regresija omogoča neposredno interpretacijo tveganja (Odds Ratio) in izpolnjuje zahtevo naloge po uporabi vsaj dveh različnih tipov regresije.

Definicija spremenljivkZa ta model smo spremenili ciljno spremenljivko in izbrali tiste napovednike, ki bi lahko vplivali na pojav napake:
* Ciljna spremenljivka ($Y$): Napake_binOpis: Binarna spremenljivka, kjer 1 pomeni, da je v koraku nastala vsaj ena napaka (popravek), in 0 pomeni, da napake ni bilo.
* Napovedna spremenljivka ($X_1$): Trajanje_hOpis: Daljše trajanje je lahko povezano z bolj kompleksnimi nalogami, ki imajo večjo verjetnost napake.
* Napovedna spremenljivka ($X_2$): Čakanje_hOpis: Čas čakanja je lahko povezan z napako (npr. daljše čakanje, ker se napaka popravlja, ali pa čakanje povzroči napako zaradi izgube konteksta).
* Napovedna spremenljivka ($X_3$): IteracijaOpis: Število ponovitev, ki že sami po sebi nakazujejo na nestabilnost ali problematičnost koraka, kar je verjetno povezano z napakami.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# 1) Ustvarimo binarno odvisno spremenljivko:
# Napake_bin = 1, če je bila v koraku napaka, sicer 0
df['Napake_bin'] = (df['Napake'] > 0).astype(int)

print("Frekvenca napak (0 = brez napake, 1 = vsaj ena napaka):")
print(df['Napake_bin'].value_counts())
print("\nDelež napak:")
print(df['Napake_bin'].value_counts(normalize=True).round(3))

# 2) Izbor napovednih spremenljivk (X) in ciljne (y)
# Vzamemo samo numerične napovednike, ki so smiselni za napoved napak
features_logit = ['Trajanje_h', 'Čakanje_h', 'Iteracija']
df_logit = df[features_logit + ['Napake_bin']].dropna()

X_log = df_logit[features_logit]
y_log = df_logit['Napake_bin']

# Dodamo konstanto za statsmodels (intercept)
X_log_const = sm.add_constant(X_log)

# 3) Logistična regresija z statsmodels
logit_model = sm.Logit(y_log, X_log_const).fit()
print(logit_model.summary())

# 4) Tabela koeficientov + odds ratio (e^β)
params = logit_model.params
conf = logit_model.conf_int()
or_vals = np.exp(params)
or_ci_low = np.exp(conf[0])
or_ci_high = np.exp(conf[1])

logit_table = pd.DataFrame({
    'β': params.round(3),
    'Odds ratio (e^β)': or_vals.round(3),
    'CI_low': or_ci_low.round(3),
    'CI_high': or_ci_high.round(3),
    'p': logit_model.pvalues.round(3)
})

print("\nTabela koeficientov in odds ratio (logistična regresija):")
print(logit_table)

# 5) McFadden pseudo-R²
pseudo_r2 = 1 - logit_model.llf / logit_model.llnull
print(f"\nMcFadden pseudo-R²: {pseudo_r2:.3f}")

# 6) 10-fold Cross-Validation za logistični model (sklearn)
log_reg = LogisticRegression(max_iter=1000, solver='lbfgs')
kf = KFold(n_splits=10, shuffle=True, random_state=42)

cv_acc = cross_val_score(log_reg, X_log, y_log, cv=kf, scoring='accuracy')
cv_auc = cross_val_score(log_reg, X_log, y_log, cv=kf, scoring='roc_auc')

print(f"\n10-fold CV accuracy: {cv_acc.mean():.3f} ± {cv_acc.std():.3f}")
print(f"10-fold CV AUC: {cv_auc.mean():.3f} ± {cv_auc.std():.3f}")
print(f"Število opazovanj (N): {len(df_logit)}")

# 7) ROC krivulja za model (prileže se v del "grafična predstavitev")
from sklearn.metrics import roc_curve, auc

# naučimo model na vseh podatkih
log_reg.fit(X_log, y_log)
y_prob = log_reg.predict_proba(X_log)[:, 1]

fpr, tpr, _ = roc_curve(y_log, y_prob)
roc_auc = auc(fpr, tpr)

plt.style.use('dark_background')
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'ROC (AUC = {roc_auc:.2f})')
plt.plot([0,1], [0,1], linestyle='--', color='gray', label='Naključni model')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Model 3 – ROC krivulja (Napoved napak)')
plt.legend()
plt.tight_layout()
plt.show()

# 8) "Pomembnost" spremenljivk – stolpčni prikaz koeficientov (β)
coef_series = pd.Series(log_reg.coef_[0], index=features_logit)

plt.figure(figsize=(6,4))
coef_series.sort_values().plot(kind='barh')
plt.title('Model 3 – Koeficienti logistične regresije (Napake_bin)')
plt.xlabel('β (log-odds)')
plt.tight_layout()
plt.show()
```

Logistična regresija za napovedovanje spremenljivke Napake_bin ni stabilna in ni primerna za interpretacijo. Razlog je izrazita neuravnoteženost razredov (91 % brez napak in 9 % z napako), kar vodi v pojav quasi-complete separation. Model lahko popolnoma napove večji del podatkov, zato se koeficienti napihnejo na ekstremno visoke vrednosti, standardni odkloni postanejo ogromni, p-vrednosti niso smiselne, pri cross-validaciji pa v nekaterih foldih ni nobenega primera razreda “1”, zato AUC ni definirana.

Tak model se matematično ne more pravilno naučiti parametrov, zato ga zavrnemo kot neprimernega za analizo procesa.

## MODEL 4 – Ordinalna regresija

Cilj Modela 4 je izpolniti zahtevo po uporabi tretjega tipa regresije in hkrati preusmeriti pogled na to, ali je mogoče trajanje koraka uspešno kategorizirati in napovedati na podlagi metrik nestabilnosti in neučinkovitosti. Uporabljamo ordinalno regresijo, ki je primerna za modeliranje kategorij, ki imajo naraven, urejen vrstni red.

Definicija spremenljivkZa razliko od prejšnjih modelov, kjer je bilo Trajanje napovednik ($X$), ga tukaj spremenimo v urejeno ciljno spremenljivko ($Y$):
* Ciljna spremenljivka ($Y$): Trajanje_kategorijaOpis: Nova, ordinalna spremenljivka, ustvarjena s kategorizacijo zvezne spremenljivke Trajanje_h v tri urejene ravni: Kratko (0), Srednje (1) in Dolgo (2).
* Napovedna spremenljivka ($X_1$): Čakanje_hOpis: Čas čakanja (ura).
* Napovedna spremenljivka ($X_2$): NapakeOpis: Število napak.
* Napovedna spremenljivka ($X_3$): IteracijaOpis: Število ponovitev koraka.


```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.miscmodels.ordinal_model import OrderedModel
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

plt.style.use("dark_background") # dark theme za vse grafe

# Predpostavka: df je DataFrame s potrebnimi stolpci

# 1) Poskrbimo, da imamo Trajanje_kategorija (Kratko / Srednje / Dolgo)
if 'Trajanje_kategorija' not in df.columns:
    def kategoriziraj_trajanje(x):
        # POZOR: Pragi (1.0 in 2.0) so vzeti kot primer
        if x < 1.0:
            return 'Kratko'
        elif x < 2.0:
            return 'Srednje'
        else:
            return 'Dolgo'

    df['Trajanje_kategorija'] = df['Trajanje_h'].apply(kategoriziraj_trajanje)

# 2) Pripravimo podatke za ordinalni model
df_ord = df[['Trajanje_kategorija', 'Čakanje_h', 'Napake', 'Iteracija']].dropna().copy()

# urejeni faktor
cat_type = pd.CategoricalDtype(categories=['Kratko', 'Srednje', 'Dolgo'], ordered=True)
df_ord['Trajanje_kategorija'] = df_ord['Trajanje_kategorija'].astype(cat_type)

# Y kot KODE 0,1,2 (ordinalno)
y4 = df_ord['Trajanje_kategorija'].cat.codes   # 0 = Kratko, 1 = Srednje, 2 = Dolgo
X4 = df_ord[['Čakanje_h', 'Napake', 'Iteracija']]

# 3) Ordinalni model (logit link)
ord_model = OrderedModel(y4, X4, distr='logit')
ord_res = ord_model.fit(method='bfgs')
print(ord_res.summary())

# Koeficienti (β), SE, p
params = ord_res.params
bse = ord_res.bse
pvalues = ord_res.pvalues

print("\nKoeficienti napovednikov:")
for name in X4.columns:
    print(f"{name}: β = {params[name]:.3f}, SE = {bse[name]:.3f}, p = {pvalues[name]:.3f}")


# --- Metrike na celotnem naboru (Train) ---
# ord_res.predict(X4) vrne DataFrame verjetnosti
pred_prob_train = ord_res.predict(X4)

# Kličemo .values, ker je pred_prob_train DataFrame!
pred_class_train = pred_prob_train.values.argmax(axis=1)

accuracy_train = accuracy_score(y4.values, pred_class_train)
print(f"\nTrain natančnost (Accuracy): {accuracy_train:.3f}")


# 4) 10-fold cross-validation za natančnost (accuracy)
kf = KFold(n_splits=10, shuffle=True, random_state=42)
X4_values = X4.values
y_codes = y4.values   # 0,1,2

acc_scores = []

for train_idx, test_idx in kf.split(X4_values):
    X_train, X_test = X4_values[train_idx], X4_values[test_idx]
    y_train, y_test = y_codes[train_idx], y_codes[test_idx]

    mdl = OrderedModel(y_train, X_train, distr='logit')
    res = mdl.fit(method='bfgs', disp=False)

    pred_prob = res.predict(X_test)          # Vrne se kot NumPy array, ker so X_test NumPy array

    # POPRAVEK: Odstranimo .values, ker je pred_prob že NumPy array
    pred_class = pred_prob.argmax(axis=1)    # indeks največje verjetnosti (0,1,2)

    acc = accuracy_score(y_test, pred_class)
    acc_scores.append(acc)


print(f"\n10-fold CV accuracy (ordinalni model): {np.mean(acc_scores):.3f} ± {np.std(acc_scores):.3f}")


# --- POSODOBLJENA IN DODANA GRAFIKA ---

# 5) Graf 1: Pomembnost spremenljivk (koeficienti β)
coef_ord = params[X4.columns]

plt.figure(figsize=(6, 4))
coef_ord.sort_values(ascending=True).plot(kind='barh', color="gold")
plt.title('Model 4: Koeficienti ordinalne regresije (Trajanje)')
plt.xlabel('β (Log-odds za prehod v višjo kategorijo)')
plt.axvline(0, color='white', linestyle='--', alpha=0.5)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
print("Komentar: Koeficient kaže, ali povečanje X poveča (pozitiven β) ali zmanjša (negativen β) verjetnost, da bo Trajanje v višji kategoriji.")


# 6) Graf 2: Porazdelitev dejanskih vs. napovedanih razredov (Train)
categories = ['Kratko (0)', 'Srednje (1)', 'Dolgo (2)']

plt.figure(figsize=(8, 4))
plt.hist([y4.values, pred_class_train], bins=np.arange(4)-0.5,
         label=['Dejansko', 'Napovedano'],
         alpha=0.6, density=False, color=['dodgerblue', 'salmon'])

plt.xticks(np.arange(3), categories)
plt.title('Model 4: Dejanska vs. napovedana porazdelitev kategorij')
plt.xlabel('Kategorija trajanja')
plt.ylabel('Število opazovanj')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
print("Komentar: Graf prikazuje, kako se napovedana porazdelitev ujema z dejansko porazdelitvijo treh kategorij.")


# 7) Graf 3: Boxplot cross-validation natančnosti (Accuracy)
plt.figure(figsize=(5, 4))
plt.boxplot(acc_scores, vert=True, patch_artist=True, boxprops=dict(facecolor='mediumpurple'))
plt.title('Model 4: 10-fold CV Natančnost (Accuracy)')
plt.ylabel('Natančnost (%)')
plt.xlabel('Model')
plt.xticks([1], ['Ordinalna regresija'])
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
print(f"Komentar: Boxplot prikazuje stabilnost natančnosti med 10 foldi. Povprečje: {np.mean(acc_scores):.3f}")


# 8) Graf 4: CV Natančnost po posameznih foldih
plt.figure(figsize=(6, 4))
plt.bar(np.arange(1, 11), acc_scores, color="rebeccapurple")
plt.title('Model 4: Natančnost po posameznih foldih (10-fold CV)')
plt.xlabel('Fold')
plt.ylabel('Natančnost (Accuracy)')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
print("Komentar: Stolpčni graf prikazuje variacijo natančnosti med 10 foldi, kar je ključno za oceno stabilnosti modela.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import (
    r2_score,
    mean_squared_error,
    mean_absolute_error,
    accuracy_score,
)

from statsmodels.miscmodels.ordinal_model import OrderedModel

plt.style.use("dark_background")

# ============================================================
# PRIPRAVA PODATKOV (skupna za vse modele)
# ============================================================

# 1) poskrbimo, da so tipi pravilni
df_lin = df[['Trajanje_h', 'Čakanje_h', 'Napake', 'Iteracija', 'Tip_aktivnosti']].dropna().copy()

# ordinalna kategorija trajanja (če še ni)
if 'Trajanje_kategorija' not in df.columns:
    def kategoriziraj_trajanje(x):
        if x < 1.0:
            return 'Kratko'
        elif x < 2.0:
            return 'Srednje'
        else:
            return 'Dolgo'
    df['Trajanje_kategorija'] = df['Trajanje_h'].apply(kategoriziraj_trajanje)

# binarna napaka (za logistični model)
df['Napake_bin'] = (df['Napake'] > 0).astype(int)

# ============================================================
# MODEL 1 – LINEARNA REGRESIJA: Čakanje_h ~ Trajanje_h
# ============================================================

X1 = df_lin[['Trajanje_h']]
y1 = df_lin['Čakanje_h']

X1_const = sm.add_constant(X1)
model1 = sm.OLS(y1, X1_const).fit()
print("=== MODEL 1 – OLS: Čakanje_h ~ Trajanje_h ===")
print(model1.summary())

# napovedi in ostanki
y1_pred = model1.predict(X1_const)
resid1 = y1 - y1_pred

# 10-fold CV (R², RMSE, MAE)
kf = KFold(n_splits=10, shuffle=True, random_state=42)
lin1 = LinearRegression()
cv_r2_m1 = cross_val_score(lin1, X1.values, y1.values, cv=kf, scoring='r2')
cv_rmse_m1 = -cross_val_score(lin1, X1.values, y1.values, cv=kf, scoring='neg_root_mean_squared_error')
cv_mae_m1 = -cross_val_score(lin1, X1.values, y1.values, cv=kf, scoring='neg_mean_absolute_error')

print(f"Model 1 – 10-fold CV R²: {cv_r2_m1.mean():.3f} ± {cv_r2_m1.std():.3f}")
print(f"Model 1 – 10-fold CV RMSE: {cv_rmse_m1.mean():.3f} ± {cv_rmse_m1.std():.3f}")
print(f"Model 1 – 10-fold CV MAE: {cv_mae_m1.mean():.3f} ± {cv_mae_m1.std():.3f}")

# ---- GRAFI MODEL 1 ----

# 1a) dejanske vs napovedane vrednosti
plt.figure(figsize=(6,4))
plt.scatter(y1, y1_pred, alpha=0.7)
plt.plot([y1.min(), y1.max()], [y1.min(), y1.max()], 'r--', label='idealno ujemanje')
plt.title("Model 1 – Dejanske vs napovedane vrednosti (Čakanje_h)")
plt.xlabel("Dejanske vrednosti")
plt.ylabel("Napovedane vrednosti")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 1b) ostanki vs napovedi
plt.figure(figsize=(6,4))
plt.scatter(y1_pred, resid1, alpha=0.7)
plt.axhline(0, color='r', linestyle='--')
plt.title("Model 1 – Ostanki vs napovedi")
plt.xlabel("Napovedane vrednosti")
plt.ylabel("Ostanki")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 1c) boxplot 10-fold R²
plt.figure(figsize=(4,4))
plt.boxplot(cv_r2_m1, vert=True)
plt.title("Model 1 – 10-fold CV R²")
plt.ylabel("R²")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================
# MODEL 2 – VEČKRATNA LINEARNA REGRESIJA
# Čakanje_h ~ Trajanje_h + Napake + Iteracija
# ============================================================

X2 = df_lin[['Trajanje_h', 'Napake', 'Iteracija']]
y2 = df_lin['Čakanje_h']

X2_const = sm.add_constant(X2)
model2 = sm.OLS(y2, X2_const).fit()
print("\n=== MODEL 2 – OLS: Čakanje_h ~ Trajanje_h + Napake + Iteracija ===")
print(model2.summary())

y2_pred = model2.predict(X2_const)
resid2 = y2 - y2_pred

# 10-fold CV za model 2 (R², RMSE, MAE)
lin2 = LinearRegression()
cv_r2_m2 = cross_val_score(lin2, X2.values, y2.values, cv=kf, scoring='r2')
cv_rmse_m2 = -cross_val_score(lin2, X2.values, y2.values, cv=kf, scoring='neg_root_mean_squared_error')
cv_mae_m2 = -cross_val_score(lin2, X2.values, y2.values, cv=kf, scoring='neg_mean_absolute_error')

print(f"Model 2 – 10-fold CV R²: {cv_r2_m2.mean():.3f} ± {cv_r2_m2.std():.3f}")
print(f"Model 2 – 10-fold CV RMSE: {cv_rmse_m2.mean():.3f} ± {cv_rmse_m2.std():.3f}")
print(f"Model 2 – 10-fold CV MAE: {cv_mae_m2.mean():.3f} ± {cv_mae_m2.std():.3f}")

# standardizirani koeficienti (β_std)
y2_std = (y2 - y2.mean()) / y2.std(ddof=1)
X2_std = (X2 - X2.mean()) / X2.std(ddof=1)
X2_std_const = sm.add_constant(X2_std)
model2_std = sm.OLS(y2_std, X2_std_const).fit()
beta_std_m2 = model2_std.params.drop('const')

# ---- GRAFI MODEL 2 ----

# 2a) dejanske vs napovedane
plt.figure(figsize=(6,4))
plt.scatter(y2, y2_pred, alpha=0.7)
plt.plot([y2.min(), y2.max()], [y2.min(), y2.max()], 'r--')
plt.title("Model 2 – Dejanske vs napovedane vrednosti (Čakanje_h)")
plt.xlabel("Dejanske vrednosti")
plt.ylabel("Napovedane vrednosti")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 2b) ostanki vs napovedi
plt.figure(figsize=(6,4))
plt.scatter(y2_pred, resid2, alpha=0.7)
plt.axhline(0, color='r', linestyle='--')
plt.title("Model 2 – Ostanki vs napovedi")
plt.xlabel("Napovedane vrednosti")
plt.ylabel("Ostanki")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 2c) pomembnost spremenljivk – standardizirani β
plt.figure(figsize=(6,4))
beta_std_m2.plot(kind='bar')
plt.title("Model 2 – Standardizirani koeficienti (β_std)")
plt.ylabel("β_std")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 2d) boxplot 10-fold R²
plt.figure(figsize=(4,4))
plt.boxplot(cv_r2_m2, vert=True)
plt.title("Model 2 – 10-fold CV R²")
plt.ylabel("R²")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================
# MODEL 3 – LOGISTIČNA REGRESIJA
# Napake_bin (0/1) ~ Trajanje_h + Čakanje_h + Iteracija
# ============================================================

df_log = df[['Trajanje_h', 'Čakanje_h', 'Iteracija', 'Napake_bin']].dropna().copy()
X3 = df_log[['Trajanje_h', 'Čakanje_h', 'Iteracija']].values
y3 = df_log['Napake_bin'].values

log_reg = LogisticRegression(solver='liblinear')
log_reg.fit(X3, y3)

print("\n=== MODEL 3 – Logistična regresija: Napake_bin ~ Trajanje_h + Čakanje_h + Iteracija ===")
print("Koeficienti (β):", log_reg.coef_[0])
print("Intercept:", log_reg.intercept_[0])

# odds ratio
odds_ratio = np.exp(log_reg.coef_[0])
print("Odds ratio (e^β):", odds_ratio)

# 10-fold CV accuracy
cv_acc_m3 = cross_val_score(log_reg, X3, y3, cv=kf, scoring='accuracy')
print(f"Model 3 – 10-fold CV accuracy: {cv_acc_m3.mean():.3f} ± {cv_acc_m3.std():.3f}")

# ---- GRAFI MODEL 3 ----

# 3a) pomembnost – odds ratio
plt.figure(figsize=(6,4))
plt.bar(['Trajanje_h', 'Čakanje_h', 'Iteracija'], odds_ratio)
plt.title("Model 3 – Odds ratio po napovednikih")
plt.ylabel("Odds ratio (e^β)")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 3b) boxplot accuracy
plt.figure(figsize=(4,4))
plt.boxplot(cv_acc_m3, vert=True)
plt.title("Model 3 – 10-fold CV accuracy")
plt.ylabel("Accuracy")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================
# MODEL 4 – ORDINALNA REGRESIJA
# Y = Trajanje_kategorija (Kratko/Srednje/Dolgo)
# X = Čakanje_h, Napake, Iteracija
# ============================================================

df_ord = df[['Trajanje_kategorija', 'Čakanje_h', 'Napake', 'Iteracija']].dropna().copy()
cat_type = pd.CategoricalDtype(categories=['Kratko', 'Srednje', 'Dolgo'], ordered=True)
df_ord['Trajanje_kategorija'] = df_ord['Trajanje_kategorija'].astype(cat_type)

y4_cat = df_ord['Trajanje_kategorija']
y4_codes = y4_cat.cat.codes  # 0,1,2
X4 = df_ord[['Čakanje_h', 'Napake', 'Iteracija']]

ord_model = OrderedModel(y4_codes, X4, distr='logit')
ord_res = ord_model.fit(method='bfgs')
print("\n=== MODEL 4 – Ordinalna regresija: Trajanje_kategorija ~ Čakanje_h + Napake + Iteracija ===")
print(ord_res.summary())

params4 = ord_res.params
bse4 = ord_res.bse
pvals4 = ord_res.pvalues

print("\nKoeficienti napovednikov (Model 4):")
for name in X4.columns:
    print(f"{name}: β = {params4[name]:.3f}, SE = {bse4[name]:.3f}, p = {pvals4[name]:.3f}")

# 10-fold CV accuracy za ordinalni model
X4_values = X4.values
acc_scores_m4 = []

for train_idx, test_idx in kf.split(X4_values):
    X_train, X_test = X4_values[train_idx], X4_values[test_idx]
    y_train, y_test = y4_codes[train_idx], y4_codes[test_idx]

    mdl4 = OrderedModel(y_train, X_train, distr='logit')
    res4 = mdl4.fit(method='bfgs', disp=False)

    pred_prob = res4.predict(X_test)           # (n, 3)
    pred_class = pred_prob.argmax(axis=1)      # 0/1/2
    acc = accuracy_score(y_test, pred_class)
    acc_scores_m4.append(acc)

print(f"\nModel 4 – 10-fold CV accuracy: {np.mean(acc_scores_m4):.3f} ± {np.std(acc_scores_m4):.3f}")

# ---- GRAFI MODEL 4 ----

# 4a) koeficienti napovednikov
coef_ord = params4[X4.columns]

plt.figure(figsize=(6,4))
coef_ord.plot(kind='bar')
plt.title("Model 4 – Koeficienti napovednikov")
plt.ylabel("β")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 4b) boxplot accuracy
plt.figure(figsize=(4,4))
plt.boxplot(acc_scores_m4, vert=True)
plt.title("Model 4 – 10-fold CV accuracy")
plt.ylabel("Accuracy")
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
